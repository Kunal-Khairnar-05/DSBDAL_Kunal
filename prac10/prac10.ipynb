```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Iris.csv")
df.head()
```

```python
df = df.drop('Id', axis=1)
df.info()
```

### 1. Features and their types

Based on the `df.info()` output, the dataset contains the following features:

*   **SepalLengthCm**: Numeric (float64)
*   **SepalWidthCm**: Numeric (float64)
*   **PetalLengthCm**: Numeric (float64)
*   **PetalWidthCm**: Numeric (float64)
*   **species**: Nominal/Categorical (object)

*   **SepalLengthCm**: The distribution appears fairly symmetric with no significant outliers.
*   **SepalWidthCm**: This feature shows a few potential outliers on both the lower and upper ends, indicating some flowers have unusually narrow or wide sepal widths compared to the majority.
