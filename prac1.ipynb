```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
print("Information : ")
df.info()
print("Description : ")
df.describe()
```

```python
df.shape
```

```python
print("Null Values  : ")
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

df.isnull().sum()
```

```python
df['PClass'] = df['PClass'].fillna(df['PClass'].mode()[0])

df.isnull().sum()
```

```python
print("Categorical to numeric : ")
df = pd.get_dummies(df, columns=["Sex"], dtype=int)
df.head()
```
