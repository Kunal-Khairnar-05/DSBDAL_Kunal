```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
print("Missing Values : ")
print(titanic.isnull().sum())
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic.isnull().sum()
```

```python
sns.countplot(x='survived', data= titanic)
plt.title("Surviced Count")
plt.show()
```

```python
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title("Survived Count by gender")
plt.show()
```

```python
sns.countplot(x='survived', hue='pclass', data=titanic)
plt.title("Survived Count by Pclass")
plt.show()
```

```python
sns.histplot(titanic['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=titanic, x='age', hue='survived', kde=True)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
sns.boxplot(x='survived', y='fare', data=titanic)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(titanic['fare'], bins=30)
plt.title("Fare Distribution by Survival")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```
