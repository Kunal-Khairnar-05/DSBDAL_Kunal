```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(titanic.corr(numeric_only=True), annot=True)
plt.title("Correlation pattern")
plt.show()
```

```python
sns.barplot(x='pclass', y='survived', data=titanic)
```

```python
plt.figure(figsize=(12,12))
```

```python
plt.subplot(3,2,1)

sns.histplot(titanic['fare'], kde=False, bins=30)
plt.title("Fare Histplot")
```

```python
plt.subplot(3,2,2)

sns.histplot(titanic['fare'], kde=True, bins=30)
plt.title("Fare Histplot With KDE")
```

```python
plt.subplot(3,2,3)

sns.histplot(titanic['fare'], kde=True, stat="density", color="orange")
plt.title("Distplot")
```

```python
plt.subplot(3,2,4)

sns.rugplot(titanic['fare'], color='red')
sns.histplot(titanic['fare'], kde=True, color="orange")
plt.title("Rugplot")
```

```python
plt.subplot(3,2,4)

sns.barplot(data=titanic, x='pclass', y='fare',color='red')

plt.title("barplot")


plt.tight_layout()

plt.show()
```
