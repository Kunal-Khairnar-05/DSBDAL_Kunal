```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['mathscore'] = df['mathscore'].fillna(df['mathscore'].mean())
df.isnull().sum()
```

```python
initial_skew = df['mathscore'].skew()
print(f"{initial_skew:.2f}")
```

```python
df['math_log'] = np.log1p(df['mathscore'])
print(df['math_log'].skew())
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1, 2 , 1)
sns.histplot(df['mathscore'], kde=True, color='royalblue', bins=20)
plt.title("Original Skew Value:")
plt.xlabel("Score")
plt.ylabel("Frequency")



plt.subplot(1, 2, 2)
sns.histplot(df['math_log'], kde=True, color="darkorange", bins=20)
plt.title("Log Skew Value:")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
