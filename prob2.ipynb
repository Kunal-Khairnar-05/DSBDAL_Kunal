```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isnull().sum()
```

```python
print("Datatypes before conversion : ", df.dtypes)
```

```python
df['Survived'] = df['Survived'].astype(bool)

df.dtypes
```

```python
# Normalizing
```

```python
def normalize(col):
    return (col - col.min()) / (col.max() - col.min())
```

```python
df['Age'] = normalize(df['Age'])

print("Normalized age : ", df['Age'].head())
```
